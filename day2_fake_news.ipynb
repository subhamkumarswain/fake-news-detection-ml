```python
import pandas as pd

df = pd.read_csv("train.tsv.zip", sep="\t", header=None)
df.head()
```

```python
df.columns = [
    "id",
    "label",
    "statement",
    "subject",
    "speaker",
    "job",
    "state",
    "party",
    "barely_true_counts",
    "false_counts",
    "half_true_counts",
    "mostly_true_counts",
    "pants_on_fire_counts",
    "context"
]

```

```python
print(df["label"].unique())
```

```python
fake_classes = ["false", "barely-true", "pants-fire"]
real_classes = ["true", "mostly-true", "half-true"]

df["label"] = df["label"].apply(
    lambda x: 0 if x in fake_classes else 1
)
```

```python
print(df["label"].unique())
print(df["label"].value_counts())
```

```python
df = df.rename(columns={"statement": "text"})
df = df[["text", "label"]]

df.drop_duplicates(inplace=True)
df = df[df["text"].str.len() > 20]

df.to_csv("fake_news_cleaned.csv", index=False)

print("Final fake dataset size:", df.shape)
```
