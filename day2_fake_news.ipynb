```python
import pandas as pd

df = pd.read_csv("train.tsv.zip", sep="\t", header=None)
df.head()
```

```python
df.columns = [
    "id",
    "label",
    "statement",
    "subject",
    "speaker",
    "job",
    "state",
    "party",
    "barely_true_counts",
    "false_counts",
    "half_true_counts",
    "mostly_true_counts",
    "pants_on_fire_counts",
    "context"
]

df.head()
```

```python
fake_labels = ["false", "pants-fire"]

df_fake = df[df["label"].isin(fake_labels)]

df_fake = df_fake[["statement"]]
df_fake.rename(columns={"statement": "text"}, inplace=True)

df_fake["label"] = 0   # 0 = FAKE

print("Total FAKE news articles:", len(df_fake))
df_fake.head()
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-z\s]", "", text)
    return text

df_fake["text"] = df_fake["text"].apply(clean_text)
df_fake.dropna(inplace=True)
```

```python
df_fake.to_csv("fake_news_cleaned.csv", index=False)
print("Saved fake_news_cleaned.csv")
```
