```python
from google.colab import userdata
API_KEY = userdata.get("NEWS_API_KEY")
```

```python
import requests
import pandas as pd



categories =["election", "government", "economy", "technology", "health"]
all_articles = []

for category in categories:

    url = f"https://newsapi.org/v2/everything?q={category}&language=en&sortBy=publishedAt&pageSize=100&apiKey={API_KEY}"
    response = requests.get(url)
    data = response.json()

    if data["status"] != "ok":
      print("Error:", data)
      break

    for article in data["articles"]:
      if article["title"] is not None:
        all_articles.append(article["title"])

df = pd.DataFrame(all_articles, columns=["text"])
df["label"] = 1   # Real news label

print("Total real news collected:", df.shape)

df.to_csv("real_news_big.csv", index=False)

```

```python
df.drop_duplicates(inplace=True)
df = df[df["text"].str.len() > 20]
```
